```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,20)
y = x*x
z = y+x
```

```python
plt.plot(x,y)
```

```python
fig = plt.figure(figsize=(3,3))
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('tete')
axes.set_ylabel('tete')
axes.set_title('tete')
axes.plot(x,y)
```

```python
fig , axes = plt.subplots(1,2,figsize=(10,5))

axes[0].plot(x,y)
axes[0].set_xlabel('tete')
axes[0].set_ylabel('tete')
axes[0].set_title('tete')

axes[1].plot(y,x)
axes[1].set_xlabel('naruto')
axes[1].set_ylabel('naruto')
axes[1].set_title('naruto')
```
