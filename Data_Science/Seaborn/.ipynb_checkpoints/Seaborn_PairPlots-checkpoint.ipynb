```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
tips= sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.residplot(x='total_bill', y='tip', data=tips, lowess=True)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', color='blue') #Kind = "value" muda a forma do grafico (hex, kde, reg)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg', color='red') #Kind = "value" muda a forma do grafico (hex, kde, reg)
```
