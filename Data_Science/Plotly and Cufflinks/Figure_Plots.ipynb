```python
import numpy as np
import pandas as pd
%matplotlib inline
import plotly as pl
import plotly.offline as po
import cufflinks as cf
```

```python
#Poder utilizar em modo offline
po.init_notebook_mode(connected=True) 
cf.go_offline()
```
